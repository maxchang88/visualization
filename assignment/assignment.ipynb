````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1

college = pd.read_csv('/content/visualization/assignment/data/college_completion.csv')

# 2

print(college.shape, '\n') # 3798 rows
college.head()
````

````python
# 3

pd.crosstab(college['control'], college['level'])
````

````python
# 4

# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(college['grad_100_value'], bins=30, kde=False, edgecolor='black')
plt.xlabel('grad_100_value')
plt.ylabel('Frequency')
plt.title('Histogram of grad_100_value')
plt.show()

# Kernel Density Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(college['grad_100_value'], fill=True)
plt.xlabel('grad_100_value')
plt.ylabel('Density')
plt.title('Kernel Density Estimate of grad_100_value')
plt.show()

# Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x=college['grad_100_value'])
plt.xlabel('grad_100_value')
plt.title('Boxplot of grad_100_value')
plt.show()

# Statistical Description
print("Statistical Summary of grad_100_value:")
print(college['grad_100_value'].describe())
````

````python
# 5

# Kernel Density Plot by Control
plt.figure(figsize=(8, 5))
sns.kdeplot(data=college, x='grad_100_value', hue='control', fill=True, common_norm=False)
plt.xlabel('grad_100_value')
plt.ylabel('Density')
plt.title('Kernel Density Plot of Graduation Rates by Control')
plt.legend(title='Control', labels=college['control'].unique())
plt.show()

# Kernel Density Plot by Level
plt.figure(figsize=(8, 5))
sns.kdeplot(data=college, x='grad_100_value', hue='level', fill=True, common_norm=False)
plt.xlabel('grad_100_value')
plt.ylabel('Density')
plt.title('Kernel Density Plot of Graduation Rates by Level')
plt.legend(title='Level', labels=college['level'].unique())
plt.show()

# Grouped Statistical Description of grad_100_value by Level and Control
grouped_stats = college.groupby(['level', 'control'])['grad_100_value'].describe()
print("\nGrouped Statistical Summary of grad_100_value by Level and Control:")
print(grouped_stats)
````

It seems that colleges in general struggle to graduate at a high rate. However, very notably Private for-profit universities seem to be the biggest perpetrator of of this issue. Interestingly, Private not for-profit universities have the best distribution of graduation rate, although it's mean is still very low being around 40%.

````python
# 6

college['levelXcontrol'] = college['level'] + ', ' + college['control']

# Grouped Kernel Density Plot by 'levelXcontrol'
plt.figure(figsize=(8, 5))
sns.kdeplot(data=college, x='grad_100_value', hue='levelXcontrol', fill=True, common_norm=False)
plt.xlabel('grad_100_value')
plt.ylabel('Density')
plt.title('Grouped Kernel Density Plot by Level and Control')
plt.legend(title='Level + Order', labels=college['levelXcontrol'].unique())
plt.show()

# Grouped Statistical Description of grad_100_value by 'levelXcontrol'
grouped_stats = college.groupby('levelXcontrol')['grad_100_value'].describe()

print("\nGrouped Statistical Summary of grad_100_value by Level and Control:")
print(grouped_stats)
````

2-year, Private not-for-profit and 4-year, Private not-for-profit universities seem to have the best graduation rates.

3. The **hazard rate** gives the instantaneous risk of an event occurring, given that it hasn't already happened. It is defined as
$$
h(t) = \dfrac{f(t)}{1-F(t)}
$$
where $F(t)$ is the distribution of arrival times and $f(t)=F'(t)$ is the density of arrival times. Roughly, conditional on $T \ge t$, what is the probability that $T=t$? The cumulative hazard is given by
$$
H(t) = \int_0^t h(z) dz.
$$
The following function computes the hazard rate and cumulative hazard using something called the Nelson-Ahlen estimator. Run your data through it. Interpret the plots for the hazard rate and cumulative hazard rate. What patterns do you see?

```
def hazard(df):
    """ Compute cumulative hazard rate using Nelson-Ahlen estimator. """
    at_risk = df.shape[0]
    arrivals = df['survival'].sort_values().dropna().tolist()
    hazard_rate = []
    times = []
    for t in arrivals:
        select = (df['survival']==t)*(df['Overall Survival Status']=='Deceased')
        deaths_t = df[select].shape[0]
        if deaths_t > 0:
            hazard_rate.append(deaths_t/at_risk)
            times.append(t)
            at_risk -= deaths_t
    cumulative_hazard = np.cumsum(hazard_rate)
    return cumulative_hazard, times, hazard_rate
```
